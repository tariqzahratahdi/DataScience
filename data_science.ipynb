# Data science

The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it.

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')

plt.title("Normal distribution")
plt.show()
```

# Load CSV file from Github
This is the easiest way to upload a CSV file in Colab. For this go to the dataset in your GitHub repository, and then click on “Raw”. Copy the URL link to the raw dataset and pass it as a parameter to the read_csv() in pandas to get the dataframe.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/tariqzahratahdi/hello-world/refs/heads/master/top_500_companies_in_the_world.csv'
df = pd.read_csv(url)
print(df)
```
