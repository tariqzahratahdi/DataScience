# Histogramme 2D

Un histogramme 2D, également appelé histogramme bivarié, permet de visualiser la relation entre deux variables numériques lorsque le nombre d'observations est important.

Le graphique est divisé en plusieurs compartiments ou zones (bins) colorées en fonction du nombre de points qu'elles contiennent.

La fonction **`pyplot.2dhist()`** permet de créer un histogramme 2D.

**Exemple:**

```python
import numpy as np
import matplotlib.pyplot as plt

# making data
np.random.seed(1)
x = np.random.normal(size = 10000)
y = x + np.random.normal(size = 10000)

# setting figure
plt.hist2d(x, y, bins = 20)

# show color bar
plt.colorbar()

# plotting figure
plt.show()
```

## Palette de couleurs

La palette de couleurs par défaut est `viridis`, mais vous pouvez la modifier avec l'argument `cmap`.

**Exemple:**

```python
plt.hist2d(x, y, bins = 20, cmap = 'BuPu')

plt.show()
```

## Nombre de compartiments

Le nombre de compartiments peut être personnalisé à l'aide de l'argument `bins`, dont la valeur par défaut est 10.

Pour définir le nombre de compartiments, on utilise une des quatre options suivantes.

**Nombre de cases pour les deux dimensions**

Si vous passez un entier à l'argument, ce nombre sera utilisé comme nombre de compartiments pour chaque axe.

**Exemple:**

```python
plt.hist2d(x, y, bins = 20)

plt.show()
```

**Nombre de compartiments personnalisé pour chaque axe**

Vous pouvez également définir le nombre de compartiments pour chaque axe à l'aide d'un array.

Le premier élément correspond au nombre de compartiments pour l'axe des X, tandis que le second correspond au nombre de compartiments pour l'axe des Y.

**Exemple:**

```python
plt.hist2d(x, y, bins = [50, 10])

plt.show()
```

**Bornes de compartiment pour les deux dimensions**

Une autre solution consiste à définir les bornes des compartiments pour les deux dimensions en passant un array contenant les bornes.

**Exemple:**

```python
plt.hist2d(x, y, bins = np.arange(-3, 3, 0.2))

plt.show()
```

**Bornes de compartiment pour chaque dimension**

La dernière option consiste à définir les bornes des compartiments pour chaque dimension.

```python
plt.hist2d(x, y, bins = (np.arange(-3, 3, 0.2), np.arange(-3, 3, 0.2)))

plt.show()
```
