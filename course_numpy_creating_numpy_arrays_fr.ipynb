# Créer des arrays NumPy

# Attributs d'un array

Les attributs d'un array sont des propriétés qui fournissent des informations sur la dimension, la forme, la taille, le type de données du array, etc.

Dans Numpy, les dimensions d'un array sont appelées axes.

<br> <br>
course-numpy-creating-numpy-arrays-01.png
<br> <br>

Les attributs du array `ndim` et `shape` renvoient respectivement la dimension et la forme du array respectivement.

# Création d'un array

Il existe plusieurs méthodes pour créer un array.

Un array peut être créé avec l'utilisation de divers types de données tels que des listes, des tuples, etc.

Le type du réseau résultant est déduit du type des éléments dans les séquences.

Les arrays sont construits à l'aide de la méthode **`numpy.array()`**.

> *Syntaxe*
<font size = "4">
```python
numpy.array(object, dtype=None)
```
</font>

*Paramètres*
* `objet`:`array_like`

* `dtype`:`type de données`, facultatif

**Exemple**: créer un array à 1 dimension (1D) de taille 3, composé d'éléments entiers:



````python
import numpy as np

# create array from a range
# 1-dimensional array of size 3
# composed of integer elements

arr = np.array(range(3))

print(arr)
print(type(arr))
print(arr.ndim)
print(arr.shape)
print(arr.dtype)
````

**Exemple**: créer un array à 2 dimension (2D) de forme (2, 3), composé d'éléments entiers:

````python
# create array from a range
# 2D array of shape (2, 3)
# composed of integer elements

arr = np.array([range(1,4), range(4,7)])

print(arr)
print(arr.ndim)
print(arr.shape)
````

# Modifier la forme d'un array

La méthode **`reshape()`** permet de modifier la forme d'un array.

Exemple: modifier la forme d'un array 1D vers un array 2D: mettre les nombres 1 à 9 dans une matrice 3 × 3:

````python
arr = np.array(range(1, 10))

print(arr)

mat = arr.reshape((3, 3))

print(mat)
````

Notez que pour que cela fonctionne, la taille du array initial doit correspondre à la taille du array résultant.

**Exemple**: modifier la forme un array 1D vers un array 3D: mettre les nombres 1 à 27 dans une matrice 3 × 3 × 3:

````python
arr = np.array(range(1, 28))

mat = arr.reshape((3, 3, 3))

print(mat)
````

# Méthodes supplémentaires pour la création de arrays

La création d'un array rempli de zéros ou de uns est très utile lorsque vous devez initialiser un array tant que les dimensions sont connues.

Une fois créé, vous pouvez ensuite remplir le array avec un ensemble de valeurs calculées.

Dans ce cas, le array rempli de uns ou de zéros agit simplement comme un conteneur.

## Array rempli de zéros

On définit un array de dimension donnée rempli de zéros en utilisant la fonction `numpy.zeros()`.

L'entrée est un tuple contenant les dimensions du array souhaité.

Le type de données par défaut est `float64`, mais cela peut être modifié avec le paramètre `dtype`.

> *Syntaxe*
<font size = "4">
```python
numpy.zeros(shape, dtype=float)
```
</font>

**Exemple**



````python
arr = np.zeros((3,4))

print(arr)
````

## Array rempli de uns

On définit un array de dimension donnée rempli de uns en utilisant la fonction `numpy.ones()`.

> *Syntaxe*
<font size = "4">
```python
numpy.ones(shape, dtype=float)
```
</font>

**Exemple**



````python
arr = np.ones((2,3), dtype=int)

print(arr)
````

## Array rempli d'une même valeur

La fonction `numpy.full()` crée un array de dimension donnée rempli d'une même valeur donnée.

> *Syntaxe*
<font size = "4">
```python
numpy.full(shape, fill_value)
```
</font>

**Exemple**

````python
arr = np.full((2,3), 3.14)

print(arr)
````

## Array vide

La fonction `empty()` crée un array de dimension donnée, rempli de valeurs qui seront ceux existant déjà à cet emplacement de mémoire.

> *Syntaxe*
<font size = "4">
```python
numpy.empty(shape, dtype=float)
```
</font>

**Exemple**

````python
# create 3 x 4 array of integers
arr = np.empty((3,4), dtype=int)

print(arr)
````

## Array rempli d'une séquence linéaire

La fonction `numpy.arange ()` permet de créer un array rempli d'une séquence linéaire.

> *Syntaxe*
<font size = "4">
```python
numpy.arange(start, stop, step, dtype=None)
```
</font>

Notez que la valeur d'argument `d'arrêt` est exclue de la plage.

**Exemple**

````python
# create an array filled with a linear sequence
# starting at 0, ending at 20, stepping by 2

arr = np.arange(0, 20, 2)

print(arr)
````

## Array rempli de valeurs équidistantes

La fonction `numpy.linspace()` permet de créer un array rempli de valeurs équidistantes sur un intervalle spécifié.

> Syntaxe
<font size = "4">
```python
numpy.linspace(start, stop, num, dtype=None)
```
</font>

*Paramètres*
* `start`: `array_like`. La valeur de départ de la séquence.
* `stop`: `array_like`. La valeur finale de la séquence.
* `num`: `int`. Le nombre de valeurs dans la séquence.
* `dtype`: `dtype`. Le type du array résultant.

**Exemple**



````python
# create an array filled with 5 values equally spaced
# between 0 and 1

arr = np.linspace(0, 1, 5)

print(arr)
````

## Array rempli de valeurs aléatoires

**Exemple**

````python
# create a 3x3 array of
# random samples from a
# uniform distribution over [0, 1)

arr = np.random.rand(3,3)

print(arr)
````

## Array rempli de valeurs aléatoires normalement distribuées

**Exemple**

````python
# create a 3x3 array of
# random samples from a normal distribution
# with mean 0 and standard deviation 1

arr = np.random.randn(3, 3)

print(arr)
````

## Array rempli d'entiers aléatoires

**Exemple**

````python
# create a 3x3 array of random integers
# in the interval [0,10)

arr = np.random.randint(0, 10, (3, 3))

print(arr)
````

# Charger un dataset à partir d'un fichier CSV

Le format CSV (Comma-separated values: valeurs séparées par des virgules) est un format de fichier texte qui utilise des virgules pour séparer les valeurs et des retours à la ligne (newline) pour séparer les enregistrements.

Un fichier CSV stocke les données tabulaires (numéros et texte) en texte brut, où chaque ligne du fichier représente généralement un enregistrement de données.

Chaque enregistrement se compose du même nombre de champs, et ceux-ci sont séparés par des virgules dans le fichier CSV.

> *Exemple de fichier CSV*
<font size = "4">
```
company,rank,revenues,profits,country
Walmart,1,485873,13643.0,USA
State Grid,2,315199,9571.3,China
Sinopec Group,3,267518,1257.9,China
China National Petroleum,4,262573,1867.5,China
Toyota Motor,5,254694,16899.3,Japan
```
</font>

## Charger un dataset

La fonction **`numpy.genfromtxt()`** permet de charger des données à partir d'un fichier texte.

> *Syntaxe*
<font size = "4">
```python
numpy.genfromtxt(fname, dtype=<class 'float'>, delimiter=None, skipheader=0)
```
</font>

*Paramètres*:
* `fname`: `file`, nom de fichier.
* `dtype`: `dtype`, facultatif, type de données du array résultant.
* `delimiter`: `str`, la chaîne utilisée pour séparer les valeurs.
* `skipheader`: `int`, facultatif, le nombre de lignes à sauter au début du fichier.

Dans l'exemple suivant, nous chargeons un fichier CSV contenant un dataset de trajets de taxis.

Les colonnes clés du dataset sont `pickup_year`, `pickup_month`, `pickup_day`, `pickup_time`, `trip_distance`, `fare_amount` et `tip_amount`.



````python
# csv file location
url = 'https://dq-content.s3.amazonaws.com/290/nyc_taxis.csv'

# Load the dataset, skipping the header row
taxi = np.genfromtxt(url, delimiter=',', skip_header=1)

# specify format for print
np.set_printoptions(formatter={'float': '{: 0.0f}'.format})

print(taxi)
````
