# Diagramme à barres

Un diagramme à barres offre une façon claire et intuitive de visualiser des données catégorielles.

Il permet de saisir les différences de taille ou de quantité entre les catégories.

Un diagramme à barres est une représentation graphique utilisant des barres rectangulaires pour comparer différentes catégories.

La hauteur ou la longueur de chaque barre correspond à la valeur qu'elle représente.

L'axe des abscisses indique généralement les catégories comparées, tandis que l'axe des ordonnées indique les valeurs associées à ces catégories.

Nous utilisons la fonction **`pyplot.bar()`** pour créer un diagramme à barres.

> *Syntaxe*
```python
pyplot.bar(x, height, width, bottom, align)
```

*Paramètres*
* `x` : les catégories.
* `height` : les valeurs correspondantes.
* `width` : la largeur des barres (par défaut : 0.8).
* `bottom` : la ligne de base des barres (par défaut : 0).
* `align` : comment aligner les barres (`center` ou `edge`).

Exemple :** visualiser les ventes de différents fruits.

````python
import matplotlib.pyplot as plt

# making data
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

# setting figure
plt.bar(fruits, sales)

# setting labels
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')

# plotting figure
plt.show()
````

# Diagramme à barres horizontales

Nous utilisons la fonction **`pyplot.barh()`** pour créer un diagramme à barres horizontales.

Cette fonction fonctionne de manière similaire à `pyplot.bar()`, mais affiche les barres horizontalement.

Exemple : Visualiser les ventes de différents fruits.

````python
# setting figure
plt.barh(fruits, sales)

# setting labels
plt.title('Fruit Sales')
plt.xlabel('Sales')
plt.ylabel('Fruits')

# plotting figure
plt.show()
````
