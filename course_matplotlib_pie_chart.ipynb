# Pie chart

# Creating pie chart plot

Pie chart is a circular chart used to display only one series of data.

The area of slices of the pie represents the percentage of the parts of the data.

The slices of pie are called *wedges*.

The **`pyplot.pie()`** method allows to create a pie chart plot.

> Syntax
<font size="4">
```python
pyplot.pie(x, labels=None)
```
</font>

*Parameters*:
* `x` : 1D array-like, the wedge sizes.
* `labels` : `list`, default: `None`, a sequence of strings providing the labels for each wedge.

**Example**

````python
import matplotlib.pyplot as plt

# setting data
categories = 'Children', 'Teens', 'Youngs', 'Elderlies'
proportions = 20, 30, 30, 20

# setting figure
plt.pie(proportions, labels=categories)

# plotting figure
plt.show()
````

## Clockwise pie chart

The pie chart is created by default counterclockwise.  
To set a clockwise direction, we set the argument `counterclock = False`.



````python
plt.pie(proportions, labels=categories, counterclock=False)

plt.show()
````

## Partial pie

If your data doesn't sum up to one and you don't want to normalize your data you can set the argument `normalize = False`, so a partial pie chart will be created.



````python
categories = 'Children', 'Teens', 'Youngs'
proportions = 0.2, 0.3, 0.3

plt.pie(proportions, labels=categories, normalize=False)

plt.show()
````
