# Graphique à secteurs

# Création d'un graphique à secteurs

Un graphique à secteurs est un graphique circulaire utilisé pour afficher une seule série de données.

L'aire des secteurs du graphique représente le pourcentage des données.

La méthode `pyplot.pie()` permet de créer un graphique à secteurs.

> Syntax
<font size="4">
```python
pyplot.pie(x, labels=None)
```
</font>

*Parameters*:
* `x` : 1D array-like, les tailles des secteurs.
* `labels` : `list`, default: `None`, une séquence de chaînes fournissant les étiquettes de chaque secteur.

**Example**

````python
import matplotlib.pyplot as plt

# setting data
categories = ('Children', 'Teens', 'Youngs', 'Elderlies')
proportions = (20, 30, 30, 20)

# setting figure
plt.pie(proportions, labels=categories)

# plotting figure
plt.show()
````

### Clockwise pie chart

The direction of rotation of the slices is counterclockwise.  
To set a clockwise direction, we set the argument `counterclock = False`.



````python
plt.pie(proportions, labels=categories, counterclock=False)

plt.show()
````

### Start angle

The first slice starts from the x-axis, so the default start angle is 0.

We use the argument `startangle` to set the start angle.

**Example:** set the argument `startangle = 90` so that the first slice starts from the y-axis.

````python
plt.pie(proportions, labels=categories, startangle=90)

plt.show()
````

### Size (radius)

We use the argument `radius` to set the size of the pie. The default value is 1.

````python
plt.pie(proportions, labels=categories, radius=0.5)

plt.show()
````

### Explode

We use the argument `explode` to explode (offset) one or some slices of the pie.  
We pass as argument an array which specifies the fraction of the radius with which to offset each slice. The length of this array is the length of the data to explode.

**Example:**

````python
plt.pie(proportions, labels=categories, explode=[0, 0, 0.1, 0])

plt.show()
````

### Display percentages

We use the argument `autopct` to display the count or the percentages for each slice.  
We pass the argument using the Python string formatting.

For example,  
`autopct = '%.1f'` will display the percentage value to 1 decimal place.  
`autopct = '%.1f%%'` will show the % symbol.

**Example:**

````python
plt.pie(proportions, labels=categories, autopct='%.1f%%')

plt.show()
````

### Fill colors

We use the argument `colors` to set the fill color for each slice.  
We pass the  argument as a sequence of colors.

````python
# setting colors sequence
color_sequence = ["#B9DDF1", "#9FCAE6", "#73A4CA", "#497AA7"]

plt.pie(proportions, labels=categories, colors=color_sequence)

plt.show()
````

### Edge color and width

We use the argument `wedgeprops` to set the color and the width of the border line of the slices.  
We pass this arguments as a `dict` of the properties `linewidth` and `edgecolor`.

````python
plt.pie(proportions, labels=categories, colors=color_sequence, wedgeprops={"linewidth": 1, "edgecolor": "white"})

plt.show()
````

## Partial pie

If your data doesn't sum up to one and you don't want to normalize your data you can set the argument `normalize = False`, so a partial pie chart will be created.



````python
groups = ('G1', 'G2', 'G3', 'G4', 'G5')
values = (0.1, 0.2, 0.1, 0.2, 0.1)

plt.pie(values, labels=groups, normalize=False)

plt.show()
````
