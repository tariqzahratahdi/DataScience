# Simple line plot

Matplotlib is a multiplatform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack.

# Importing matplotlib

We will use some standard shorthands for Matplotlib imports:

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

# Setting styles

We will use the `plt.style` directive to choose appropriate aesthetic styles for our figures.

Here we will set the classic style, which ensures that the plots we create use the classic Matplotlib style:

```python
# setting styles
plt.style.use('classic')
```

# Plotting

The function `plt.show()` starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display your figure or figures.

One thing to be aware of: `the plt.show()` command should be used only once per Python session, and is most often seen at the very end of the script. Multiple `show()` commands can lead to unpredictable backend-dependent behavior, and should mostly be avoided.

# Simple line plot

The simple line plot allows the visualization of a single function
`y = f (x)`.

Example: plot a sinusoid: `y = sin(x)`.

```python
# setting styles
plt.style.use('seaborn-v0_8-whitegrid')

# setting data
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

# setting figure
plt.plot(x, y)

# plotting figure
plt.show()
```

# Adjusting the plot: line colors and styles

The first adjustment you might wish to make to a plot is to control the line colors and styles.

The `plt.plot()` function takes additional arguments that can be used to specify these.

To adjust the color, we use the keyword argument `color` which accepts a string representing the color.

## Line colors

The color can be specified in a variety of ways.

|Color specification|`color` keyword string example|
|:---|:---|
|Specify color by name|`color='blue'`|
|Short color code (rgbcmyk)|`color='g'`|
|Grayscale between 0 and 1|`color='0.75'`|
|Hex code (RRGGBB from 00 to FF)|`color='#FFDD44'`|
|RGB tuple, values between 0 and 1|`color='1.0,0.2,0.3'`|
|All HTML color names supported|`color='chartreuse'`|

The single-character `color` code (e.g. `color='g'`) reflects the standard abbreviations in the RGB (Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, commonly used for digital color graphics.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values between 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse') # all HTML color names supported
```

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

## Line style

You can adjust the line style using the `linestyle` keyword.

|Line style|`linestyle` keyword string example|
|:---|:---|
|Solid|`linestyle='solid'` or `linestyle='-'`|
|Dashed|`linestyle='dashed'` or `linestyle='--'`|
|Dashdot|`linestyle='dashdot'` or `linestyle='-.'`|
|Dotted|`linestyle='dotted'` or `linestyle=':'`|

**Example**

```python
# setting data
x = np.linspace(0, 10, 100)

# setting figure
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

We can combine the `linestyle` and `color` codes into a single nonkeyword argument to the `plt.plot()` function.

**Example**

```python
# setting figure
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r') # dotted red
```

## Using the **`setp()`** method

A useful related method is `setp()` which allows to set multiple properties on a set of lines.

**Example:**

```python
# setting data
x = np.linspace(0, 2 * np.pi, 200)

# setting figure
lines = plt.plot(x, np.sin(x), x, np.cos(x))

# setting line properties
plt.setp(lines, linewidth=2, color='g', linestyle='--')
```

# Adjusting the plot: axes limits

Matplotlib chooses default axes limits for the plot, but allows to have finer control on axes limits.

To adjust axis limits we use the `plt.xlim()` and `plt.ylim()` methods.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.xlim(-1, 7)
plt.ylim(-1.5, 1.5)
```

A useful related method is `plt.axis()`.

This method allows you to set the `x` and `y` axes limits with a single call, by passing a list that specifies `[xmin, xmax, ymin, ymax]`.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.axis([-1, 7, -1.5, 1.5])
```

The `plt.axis()` allows also to automatically tighten the axes limits around the current plot.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.axis('tight')
```

The `plt.axis()` method allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in `x` is equal to one unit in `y`.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.axis('equal')
```

# Labeling plots

There are methods that allow to set labels for titles, axes and legends:
`plt.title()`, `plt.xlabel()`, `plt.ylabel()`, `plt.legend()`.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# setting labels
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

When multiple lines are being shown within a single axes, it can be useful to create a plot legend that labels each line type.

The `plt.legend()` method allows to create such a legend.

We can specify the label of each line using the `label` keyword of the `plt.plot()` function.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '-b', label='cos(x)')

plt.legend()

# adjust axes limits
plt.axis('equal')
```

The `plt.legend()` function keeps track of the line style and color, and matches these with the correct label.
