# Calcul sur les arrays NumPy

Les opérations sur des arrays peuvent être effectuées sur un array particulier ou une combinaison d'arrays.

Ces opérations comprennent certaines opérations mathématiques de base ainsi que des opérations unaires et binaires.


Les opérations vectorisées dans NumPy sont implémentées via des fonctions universelles (*ufuncs*), dont le but principal est d'exécuter rapidement des opérations répétées sur des valeurs dans des arrays NumPy.

# La puissance des opérations vectorisées

L'un des plus grands avantages de l'utilisation de NumPy est sa rapidité.

Les opérations vectorielles de NumPy font une différence significative lorsqu'on travaille avec de grands ensembles de données.

Commençons par regarder un exemple de boucle Python qui effectue plusieurs calculs (exponentiation à la puissance deux, trois et quatre) sur un grand ensemble de données.

````python
import time

# Create a large Python list
python_list = list(range(1000000))

# Start time
start_time = time.time()

# Perform multiple calculations using a Python loop
result = 0
for x in python_list:
    result += (x**2 + x**3 + x**4)

# End time and print the time taken
end_time = time.time()
print(f"Python Loop Time Taken: {end_time - start_time:.6f} seconds")
````

Voyons maintenant comment NumPy gère la même tâche en utilisant des opérations vectorisées.

Notez que nous utiliserons `np.float64` pour l'array NumPy afin de gérer les grandes valeurs générées en augmentant les nombres vers des puissances élevées:



````python
import numpy as np
import time

# Create a large NumPy array with float data type
NumPy_arr = np.arange(1000000, dtype=np.float64)

# Start time
start_time = time.time()

# Perform the same multiple calculations with vectorized operations
np_result = np.sum(NumPy_arr**2 + NumPy_arr**3 + NumPy_arr**4)

# End time and print the time taken
end_time = time.time()
print(f"NumPy Time Taken: {end_time - start_time:.6f} seconds")
````

Cette comparaison met en évidence la puissance des opérations vectorisées de NumPy, en particulier lors de l'exécution de calculs complexes sur des ensembles de données plus volumineux.

# Opérations entre un scalaire et un array

**Exemple**

````python
a = np.arange(1,10).reshape(3,3)
print(a)

# adding 1 to every element
print(a + 1)

# subtracting 1 from each element
print(a - 1)

# multiply each element by 2
print(a * 2)
````

Tous ces opérateurs arithmétiques sont simplement des couvertures pratiques autour de fonctions spécifiques intégrées à NumPy.

**Exemple**

````python
a = np.arange(1,10).reshape(3,3)

print(a)

# adding 1 to every element
b = np.add(a, 1)

print(b)

# subtracting 1 from each element
b = np.subtract(a, 1)

print(b)

# multiply each element by 2
b = np.multiply(a, 2)

print(b)
````

# Opérations entre deux arrays

**Exemple**

````python
a = [1, 2, 3]
b = [4, 5, 6]

c = np.add(a, b)

print(c)

c = np.subtract(a, b)

print(c)

c = np.multiply(a, b)

print(c)
````

# Opérations sur un array

## Valeur absolue

La ufunc `np.absolute` de NumPy renvoie la valeur absolue des éléments d'un array.

Elle est également disponible sous l'alias `np.abs`.

**Exemple**

````python
 a = np.negative(np.arange(1,10).reshape(3,3))

print(a)

# absolute value
b = np.absolute(a)

print(b)
````

Cette ufunc peut également gérer des données complexes et renvoie le module.

````python
a = np.array([3+4j, 3-4j, 1+0j, 2+0j])

print(a)

# absolute value
b = np.absolute(a)

print(b)
````

## Fonctions trigonométriques

Les fonctions trigonométriques sont très utiles en Data Science.

**Exemple**

````python
 a = np.linspace(0, np.pi, 3)

print(a)
print(np.sin(a))
print(np.cos(a))
print(np.tan(a))
````

Remarque: les valeurs sont calculées dans la précision de la machine, c'est pourquoi les valeurs qui devraient être nulles ne valent pas toujours exactement zéro.

## Exponentielles et logarithmes

Parmi les opérations communes disponibles comme ufunc NumPy on trouve les exponentielles et les logarithmes.

**Exemple**

````python
 a = [1, 2, 3]

print(a)
print(np.exp(a))
print(np.exp2(a))
print(np.power(3, a))

b = [1, 2, 4, 10]
print(b)
print(np.log(b))
print(np.log2(b))
print(np.log10(b))
````

# Agrégation

NumPy fournit des fonctions d'agrégation pour calculer des statistiques sommaires pour un ensemble de données.

Les statistiques sommaires les plus courantes sont la moyenne et l'écart type, qui permettent de résumer les valeurs `typiques` dans un ensemble de données.

D'autres agrégats sont également utiles, comme la somme, le produit, la médiane, le minimum, le maximum, les quantiles, etc.

## Résumé à cinq chiffres

Le résumé à cinq chiffres est un ensemble de statistiques descriptives qui fournit des informations sur un ensemble de données. Il comprend les cinq percentiles d'échantillon les plus importants: minimum, quartile inférieur, médiane, quartile supérieur et maximum.

**Exemple**: nous utilisons la fonction `np.percentile` pour calculer le résumé à cinq chiffres d'un ensemble de données.

````python
data = np.random.randint(1, 100, 10)

print(np.sort(data))

# Five-number summary
summary = np.percentile(data, [0, 25, 50, 75, 100], method="midpoint")

print(summary)
````

## Agrégats multidimensionnels

Un type commun d'agrégation est un agrégat le long d'une ligne ou d'une colonne.

**Exemple**

````python
a = np.random.randint(1, 100, (3, 3))

print(a)

# min and max within each column
print(np.min(a, axis=0))
print(np.max(a, axis=0))

# min and max within each row
print(np.min(a, axis=1))
print(np.max(a, axis=1))
````

# Broadcasting

Pour les arrays de la même taille, les opérations binaires sont effectuées élément par élément.

Le broadcasting (diffusion) permet de réaliser ces types d'opérations binaires sur des arrays de tailles différentes.

Le broadcasting est un ensemble de règles pour appliquer les unfuncs binaires (addition, soustraction, multiplication, etc.) sur des arrays de tailles différentes.

<br> <br>

<br> <br>

**Exemple**

````python
a = np.array([0, 1, 2])

print(a)
print(a + 5)
````

Nous pouvons considérer cela comme une opération qui diffuse ou duplique la valeur 5 dans le array [5, 5, 5], et ajoute les résultats.

L'avantage de la diffusion de NumPy est que cette duplication de valeurs n’a pas réellement lieu, mais c'est un modèle mental utile lorsque nous pensons à la diffusion.

Nous pouvons également étendre cela à des arrays de dimension supérieure.

**Exemple**: ajouter un array 1D à un array 2D.

````python
a = np.ones((3, 3))
b = np.arange(3)

print(a)
print(b)
print(a + b)
````

Ici, l'array 1D `b` est diffusé, à travers la deuxième dimension afin de correspondre à la forme du array 2D `a`.

**Exemple**: ajouter un array 1D à un array colonne 2D.

````python
a = np.arange(3).reshape(3,1)
b = np.arange(3)

print(a)
print(b)
print(a + b)
````

## Règles de broadcasting

La diffusion dans NumPy suit un ensemble de règles strictes pour déterminer l'interaction entre les deux arrays:

Règle 1: Si les deux arrays diffèrent par leur nombre de dimensions, la forme de celui qui a le moins de dimensions est complétée par des `1` sur son côté gauche.

Règle 2: Si les formes des deux arrays ne correspondent dans aucune dimension, l'array dont la forme est égale à `1` dans cette dimension est étiré pour correspondre à l'autre forme.

Règle 3: Si, dans une dimension, les tailles diffèrent et qu'aucune n'est égale à `1`, une erreur est générée.

**Exemple**

Considérons l'addition d'un array bidimensionnel à un array unidimensionnel:

```python
a = np.ones((2, 3))
b = np.arange(3)
```

Considérons une opération sur ces deux arrays.

Les formes des arrays sont:

`a.shape = (2, 3)`

`b.shape = (3,)`

Nous voyons par la règle 1 que le array `b` a moins de dimensions, donc nous le complétons à gauche avec des unités:

`a.shape -> (2, 3)`

`b.shape -> (1, 3)`

Selon la règle 2, nous voyons maintenant que la première dimension ne correspond pas, nous étirons donc cette dimension pour correspondre:

`a.shape -> (2, 3)`

`b.shape -> (2, 3)`

Les formes correspondent, et nous voyons que la forme finale sera (2, 3).



````python
a = np.ones((2, 3))
b = np.arange(3)

print(a)
print(b)
print(a + b)
````
