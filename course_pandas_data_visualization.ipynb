# Data Visualization

In the following we will use a dataset containing data about the population of countries.

```python
import pandas as pd

# csv file location
url = 'https://raw.githubusercontent.com/tariqzahratahdi/DataScience/refs/heads/main/datasets/population.csv'

# making dataframe from csv file
data = pd.read_csv(url)

# show dataframe
data
```

### Making a Pivot Table

**Drop null values:**

```python
# drop null values
data.dropna(inplace=True)

# show dataframe
data
```

**Make a pivot table:**

```python
# make a pivot table
data_pivot = data.pivot(index='year', columns='country', values='population')

# show dataframe
data_pivot
```

**Select data:**

```python
# filter data
data_filter = data['country'].isin(['Mauritania', 'Morocco', 'Algeria', 'Tunisia', 'Libya'])

# select data
data_selected = data[data_filter]

# show dataframe
data_selected
```

**Make a pivot table**

```python
# make a pivot table
data_pivot = data_selected.pivot(index='year', columns='country', values='population')

# show dataframe
data_pivot
```

## Data Visualization

### Line Plot

```python
# line plot
data_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population Evolution')
```

### Bar Plot

**Single Bar Plot**

Select only one year:

```python
# select only one year (2020)
data_pivot_filter = data_pivot.index.isin([2020])
data_pivot_selected = data_pivot[data_pivot_filter]

# show dataframe
data_pivot_selected
```

Transpose dataframe: columns to rows and vice versa:

```python
# transpose dataframe: columns to rows and vice versa
data_pivot_selected = data_pivot_selected.T

# show dataframe
data_pivot_selected
```

Show bar plot:

```python
# bar plot
data_pivot_selected.plot(kind='bar')
```

**Bar plot grouped by n variables**

Select a few years:

```python
# select few years
data_pivot_filter = data_pivot.index.isin([1980, 1990, 2000, 2010, 2020])
data_pivot_selected = data_pivot[data_pivot_filter]

# show dataframe
data_pivot_selected
```

Show bar plot:

```python
# bar plot
data_pivot_selected.plot(kind='bar', xlabel='Year', ylabel='Population', title='Population Evolution')
```

### Pie Chart

Selecting only one year:

```python
# select only one year (2020)
data_pivot_filter = data_pivot.index.isin([2020])
data_pivot_selected = data_pivot[data_pivot_filter]

# transpose dataframe: columns to rows and vice versa
data_pivot_selected = data_pivot_selected.T

# show dataframe
data_pivot_selected
```

Rename column:

```python
# rename column
data_pivot_selected.rename(columns={2020:'2020'}, inplace=True)

# show dataframe
data_pivot_selected
```

Show pie chart:

```python
# pie chart
data_pivot_selected.plot(kind='pie', y='2020', ylabel='Year 2020', title='Population Proportions')
```

### Box Plot

**Single Box Plot**

Select only one country:

```python
# select only one country
data_pivot_selected = data_pivot['Morocco']

# show dataframe
data_pivot_selected
```

Show box plot:

```python
# show box plot
data_pivot_selected.plot(kind='box', ylabel='Population', title='Population Distribution')
```

**Multiple Box Plots**

```python
# show multiple box plots
data_pivot.plot(kind='box', ylabel='Population', title='Population Distribution by Country')
```

### Histogram

**Single Histogram**

Show histogram:

```python
# show histogram
data_pivot_selected.plot(kind='hist', xlabel='Population', title='Population Frequency Distribution')
```

Select countries:

```python
# select countries
data_pivot_selected = data_pivot[['Morocco', 'Algeria']]

# show dataframe
data_pivot_selected
```

Show multiple histograms:
