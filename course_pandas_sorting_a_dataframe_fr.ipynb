```python
# importing pandas
import pandas as pd

# csv file location
url = 'https://dq-content.s3.amazonaws.com/291/f500.csv'

# making data frame from csv file
data = pd.read_csv(url, index_col = 'company')
```

# Tri d'un DataFrame

Nous utilisons la méthode **`sort_values()`** pour trier un `DataFrame`.

```python
# selecting data
data_selected = data[['profits']]

# sorting data
data_sorted = data_selected.sort_values('profits', ascending=False)

data_sorted
```

# Paramètres de tri multiples

Nous utilisons l'attribut **`by`** de la méthode `sort_values()` pour effectuer un tri multiple.

```python
# select data
data_selected = data[['sector', 'profits']]

# set sort attributes
sort_by = ['sector', 'profits']
sort_direction_list = [True, False]

# sort data
data_sorted = data_selected.sort_values(by=sort_by, ascending=sort_direction_list)

data_sorted
```
