# Simple line plot

Matplotlib is a multiplatform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack.

# Importing matplotlib

We will use some standard shorthands for Matplotlib imports:

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

# Setting styles

We will use the `pyplot.style` directive to choose appropriate aesthetic styles for our figures.

Here we will set the classic style, which ensures that the plots we create use the classic Matplotlib style:

```python
# setting styles
plt.style.use('classic')
```

# Plotting

The function `pyplot.plot()` allows to draw a plot.

The function `pyplot.show()` starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display your figure or figures.

# Simple line plot

The simple line plot allows the visualization of a single function
`y = f (x)`.

We will use `numpy` to create data for `x` and `y`.

```python
import numpy as np
```

**Example:** plot a sinusoid: `y = sin(x)`.

```python
# setting styles
plt.style.use('seaborn-v0_8-whitegrid')

# setting data
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

# setting figure
plt.plot(x, y)

# plotting figure
plt.show()
```

# Adjusting the plot: line colors and styles

The first adjustment you might wish to make to a plot is to control the line colors and styles.

The `pyplot.plot()` function takes additional arguments that can be used to specify these.

To adjust the color, we use the keyword argument `color` which accepts a string representing the color.

## Line colors

The color can be specified in a variety of ways.

|Color specification|`color` keyword string example|
|:---|:---|
|Specify color by name|`color='blue'`|
|Short color code (rgbcmyk)|`color='g'`|
|Grayscale between 0 and 1|`color='0.75'`|
|Hex code (RRGGBB from 00 to FF)|`color='#FFDD44'`|
|RGB tuple, values between 0 and 1|`color='1.0,0.2,0.3'`|
|All HTML color names supported|`color='chartreuse'`|

The single-character `color` code (e.g. `color='g'`) reflects the standard abbreviations in the RGB (Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, commonly used for digital color graphics.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values between 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse') # all HTML color names supported

# plotting figure
plt.show()
```

## Line style

You can adjust the line style using the `linestyle` keyword.

|Line style|`linestyle` keyword string example|
|:---|:---|
|Solid|`linestyle='solid'` or `linestyle='-'`|
|Dashed|`linestyle='dashed'` or `linestyle='--'`|
|Dashdot|`linestyle='dashdot'` or `linestyle='-.'`|
|Dotted|`linestyle='dotted'` or `linestyle=':'`|

**Example**

```python
# setting data
x = np.linspace(0, 10, 100)

# setting figure
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# plotting figure
plt.show()
```

We can combine the `linestyle` and `color` codes into a single nonkeyword argument to the `pyplot.plot()` function.

**Example**

```python
# setting figure
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r') # dotted red

# plotting figure
plt.show()
```

## Using the **`setp()`** method

A useful related method is `setp()` which allows to set multiple properties on a set of lines.

**Example:**

```python
# setting data
x = np.linspace(0, 2 * np.pi, 200)

# setting figure
lines = plt.plot(x, np.sin(x), x, np.cos(x))

# setting line properties
plt.setp(lines, linewidth=2, color='g', linestyle='--')

# plotting figure
plt.show()
```

# Adjusting the plot: axes limits

Matplotlib chooses default axes limits for the plot, but allows to have finer control on axes limits.

To adjust axis limits we use the `pyplot.xlim()` and `pyplot.ylim()` methods.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.xlim(-1, 7)
plt.ylim(-1.5, 1.5)

# plotting figure
plt.show()
```

A useful related method is `pyplot.axis()`.

This method allows you to set the `x` and `y` axes limits with a single call, by passing a list that specifies `[xmin, xmax, ymin, ymax]`.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.axis([-1, 7, -1.5, 1.5])

# plotting figure
plt.show()
```

The `pyplot.axis()` allows also to automatically tighten the axes limits around the current plot.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.axis('tight')

# plotting figure
plt.show()
```

The `pyplot.axis()` method allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in `x` is equal to one unit in `y`.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# adjust axes limits
plt.axis('equal')

# plotting figure
plt.show()
```

# Labeling plots

There are methods that allow to set labels for titles, axes and legends:
`pyplot.title()`, `pyplot.xlabel()`, `pyplot.ylabel()`, `pyplot.legend()`.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x))

# setting labels
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

# plotting figure
plt.show()
```

When multiple lines are being shown within a single axes, it can be useful to create a plot legend that labels each line type.

The `pyplot.legend()` method allows to create such a legend.

We can specify the label of each line using the `label` keyword of the `pyplot.plot()` function.

**Example**

```python
# setting figure
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '-b', label='cos(x)')

plt.legend()

# adjust axes limits
plt.axis('equal')

# plotting figure
plt.show()
```

The `pyplot.legend()` function keeps track of the line style and color, and matches these with the correct label.

## Color palette

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

Matplotlib allows modifying the default color cycle by setting color palettes.

**Example**

```python
# setting custom color cycle
custom_colors = ['#90CAF9', '#FFCC80', '#A5D6A7', '#EF9A9A']
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=custom_colors)

# setting data
x = np.linspace(0, 2 * np.pi, 200)

# setting figure
plt.plot(x, np.sin(x - 0))
plt.plot(x, np.sin(x - 1))
plt.plot(x, np.sin(x - 2))
plt.plot(x, np.sin(x - 3))

# plotting figure
plt.show()
```
